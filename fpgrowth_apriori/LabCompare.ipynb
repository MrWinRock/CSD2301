```python
from mlxtend.frequent_patterns import apriori, fpgrowth
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
data1 = pd.read_csv('generated_transactions.csv')
df1 = pd.DataFrame(data1)
df1.insert(0, 'Transaction ID', range(1, 1 + len(df1)))
df1.set_index('Transaction ID', inplace=True)
```

```python
data2 = pd.read_csv('sample_transactions_true_false.csv')
df2 = pd.DataFrame(data2)
df2.insert(0, 'Transaction ID', range(1, 1 + len(df1)))
df2.set_index('Transaction ID', inplace=True)
```

# **From Apriori**

---



**Generated Transactions**



```python
df1
```

```python
frequent_itemsets_apriori1 = apriori(df1, min_support=0.1, use_colnames=True)
```

```python
association_rules_apriori1 = association_rules(frequent_itemsets_apriori1, metric="confidence", min_threshold=0.4)
```

```python
print("Frequent Itemsets from Apriori:\n", frequent_itemsets_apriori1)
print("\nAssociation Rules from Apriori:\n", association_rules_apriori1)
```

**Sample Transactions**


```python
df2
```

```python
frequent_itemsets_apriori2 = apriori(df2, min_support=0.1, use_colnames=True)
```

```python
association_rules_apriori2 = association_rules(frequent_itemsets_apriori2, metric="confidence", min_threshold=0.5)
```

```python
print("Frequent Itemsets from Apriori:\n", frequent_itemsets_apriori2)
print("\nAssociation Rules from Apriori:\n", association_rules_apriori2)
```
