# 1. จากข้อมูลที่กำหนด ให้นักศึกษาทำการสร้าง data frame จากตารางที่กำหนด

```python
df = pd.DataFrame([['Gong Yoo', 26, 85], ['Hyun Bin', 24, 63], ['Lee Min Ho', 23, 55], ['Song Joong Ki', 22, 74], ['Lee Jong Suk', 23, 31], ['Park Bo-Gum', 24, 77], ['Lee Min Ho', 26, 85], ['Hyun Bin', 24, 63], ['Lee Dong Wook', 22, 42], ['Lee Min Ho', 23, 62], ['Kim Woo Bin', 24, 89], ['Go Soo', 24, 77]], columns=['Name', 'Age', 'Score'])
```

```python
df
```

```python
# ก
df['is_duplicate'] = df.duplicated()
# add column ชื่อ is_duplicate ใส่ค่าว่ามี ข้อมูลซ้ำในตาราง ? True : False
```

```python
# ข
df.drop_duplicates()
# ลบข้อมูลซ้ำออกจากตาราง
```

```python
# ค
df.drop_duplicates(keep='last')
# ลบข้อมูลซ้ำออกจากตาราง เอาข้อมูลล่าสุด
```

```python
# ง
result_df = df.drop_duplicates(keep=False)
result_df
# ลบข้อมูลซ้ำออกจากตาราง เอาข้อมูลที่ไม่ซ้ำกัน
```

```python
# จ
df.drop_duplicates(['Name'], keep='last')
# ลบข้อมูลซ้ำออกจากตาราง เอาข้อมูล Name ที่ล่าสุด
```

# 2. จากข้อมูลที่กำหนด ให้นักศึกษาทำการสร้าง data frame จากตารางที่กำหนด

image.png

```python
df2 = pd.DataFrame([['Gong Yoo', 'Arizona', 'Pass', 63.0], ['Hyun Bin', 'Georgia', 'Pass', 48.0], ['Lee Min Ho', 'New York', 'Fail', 56.0], ['Song Joong Ki', 'Indiana', 'Fail', 75.0], ['Lee Jong Suk', 'Florida', 'Pass', np.nan], ['Park Bo-Gum', 'California', 'Pass', 77.0], ['Ji Sung', np.nan, np.nan, np.nan], [np.nan, np.nan, np.nan, np.nan]], columns=['Name', 'State', 'Star', 'Score'])
```

```python
df2
```

```python
# ก
df2.dropna()
```

```python
# ข
df2.dropna(axis=0, how='all')
```

```python
# ค
df2.dropna(axis=0, thresh=2)
```

```python
# ง
df2['Star'].dropna(axis=0)
```

```python
# จ
df2.fillna(0)
```

```python
# ฉ
df2_mean = df2.copy()
df2_mean['Name'] = df2['Name'].fillna(df2['Name'].mode()[0])
df2_mean['State'] = df2['State'].fillna(df2['State'].mode()[0])
df2_mean['Star'] = df2['Star'].fillna(df2['Star'].mode()[0])
df2_mean['Score'] = df2['Score'].fillna(df2['Score'].mean())
df2_mean
```

```python
# ช
df2_median = df2.copy()
df2_median['Name'] = df2['Name'].fillna(df2['Name'].mode()[0])
df2_median['State'] = df2['State'].fillna(df2['State'].mode()[0])
df2_median['Star'] = df2['Star'].fillna(df2['Star'].mode()[0])
df2_median['Score'] = df2['Score'].fillna(df2['Score'].median())
df2_median
```
